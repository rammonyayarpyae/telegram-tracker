## % forwards by channel

```python
df["is_forward"].describe()
```

```python
# Create pivot table
forwards_by_channel = pd.pivot_table(df, values=["is_forward", "channel_id"], index=["channel_name"], aggfunc={"is_forward": np.sum, "channel_id": np.count_nonzero})

# Calculate % of messages that are forwards
forwards_by_channel["percentage_forwards"] = forwards_by_channel["is_forward"] / forwards_by_channel["channel_id"] * 100

# Sort dataframe by % forwards
forwards_by_channel = forwards_by_channel.sort_values(by=['percentage_forwards'], ascending=False)

display(forwards_by_channel)
```

```python
forwards_by_channel.to_csv("forwards_by_channel.csv")
```

## On average, how many posts does each channel make per day?

```python
avg_posts_per_day = pd.pivot_table(df, values=["channel_id", "day", "views"], index=["channel_name"], aggfunc={"channel_id": np.count_nonzero, "day": np.unique, "views": np.sum})
```

```python
display(avg_posts_per_day)
```

```python
avg_posts_per_day['num_days_posted'] = avg_posts_per_day['day'].str.len()
avg_posts_per_day['first_day_posted'] = avg_posts_per_day['day'].str[0]
avg_posts_per_day = avg_posts_per_day.drop(columns=['day']).rename(columns={"channel_id": "total_posts"})

avg_posts_per_day['avg_posts_per_day'] = avg_posts_per_day['total_posts'] / avg_posts_per_day['num_days_posted']
avg_posts_per_day['avg_views_per_day'] = avg_posts_per_day['views'] / avg_posts_per_day['num_days_posted']
avg_posts_per_day['views_per_post'] = avg_posts_per_day['views'] / avg_posts_per_day['total_posts']
avg_posts_per_day = avg_posts_per_day.sort_values(by=['avg_views_per_day'], ascending=False)
```

```python
avg_posts_per_day['num_days_posted'].describe()
```

```python
avg_posts_per_day.to_csv('avg_posts_and_views_per_day.csv')
```
